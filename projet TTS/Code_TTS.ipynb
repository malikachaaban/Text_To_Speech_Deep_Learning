```python
replacements = [
    ("à", "a"),
    ("á", "a"),
    ("â", "a"),
    ("ç", "c"),
    ("é", "e"),
    ("è", "e"),
    ("ê", "e"),
    ("ë", "e"),
    ("ï", "i"),
    ("î", "i"),
    ("í", "i"),
    ("ô", "o"),
    ("ö", "o"),
    ("ü", "u"),
    ("ù", "u"),
    ("û", "u"),
]

import re

def cleanup_text(inputs):
    text = inputs["text"]

    # Minuscule
    text = text.lower()

    # Remplacer les caractères accentués
    for src, dst in replacements:
        text = text.replace(src, dst)

    # Supprimer la ponctuation (sauf si le tokenizer peut la gérer)
    text = re.sub(r"[^\w\s0-9]", "", text)  # garde lettres et chiffres uniquement

    # Supprimer les doubles espaces
    text = re.sub(r"\s+", " ", text).strip()

    inputs["text"] = text
    return inputs

dataset = dataset.map(cleanup_text)
```

```python
arabizi_map = {
    "3": "ع",
    "7": "ح",
    "9": "ق",
    "2": "ء",
    "5": "خ",
    "6": "ط",
    "8": "غ"
}

def arabizi_to_arabic(text):
    for digit, arabic_char in arabizi_map.items():
        text = text.replace(digit, arabic_char)
    return text

def preprocess_example(example):
    example["text"] = arabizi_to_arabic(example["text"])
    return example

dataset = dataset.map(preprocess_example)
```

```python
import os
import torch
from speechbrain.pretrained import EncoderClassifier

spk_model_name = "speechbrain/spkrec-xvect-voxceleb"

device = "cuda" if torch.cuda.is_available() else "cpu"
speaker_model = EncoderClassifier.from_hparams(
    source=spk_model_name,
    run_opts={"device": device},
    savedir=os.path.join("/tmp", spk_model_name),
)


def create_speaker_embedding(waveform):
    with torch.no_grad():
        speaker_embeddings = speaker_model.encode_batch(torch.tensor(waveform))
        speaker_embeddings = torch.nn.functional.normalize(speaker_embeddings, dim=2)
        speaker_embeddings = speaker_embeddings.squeeze().cpu().numpy()
    return speaker_embeddings
```

```python
import numpy as np

def prepare_dataset(example):
    audio = example["audio"]

    example = processor(
        text=example["text"],
        audio_target=audio["array"],
        sampling_rate=audio["sampling_rate"],
        return_attention_mask=False,
    )

    # Strip batch dimension
    example["labels"] = example["labels"][0]
# Compute stop_labels: 0s with a final 1 (stop token)
    length = len(example["labels"])
    stop_labels = np.zeros(length, dtype=np.int64)
    stop_labels[-1] = 1
    example["stop_labels"] = stop_labels
    # Add speaker embedding
    example["speaker_embeddings"] = create_speaker_embedding(audio["array"])



    return example
```

```python
from dataclasses import dataclass
from typing import Any, Dict, List, Union


@dataclass
class TTSDataCollatorWithPadding:
    processor: Any

    def __call__(
        self, features: List[Dict[str, Union[List[int], torch.Tensor]]]
    ) -> Dict[str, torch.Tensor]:
        input_ids = [{"input_ids": feature["input_ids"]} for feature in features]
        label_features = [{"input_values": feature["labels"]} for feature in features]
        speaker_features = [feature["speaker_embeddings"] for feature in features]

        # collate the inputs and targets into a batch
        batch = processor.pad(
            input_ids=input_ids, labels=label_features, return_tensors="pt"
        )

        # replace padding with -100 to ignore loss correctly
        batch["labels"] = batch["labels"].masked_fill(
            batch.decoder_attention_mask.unsqueeze(-1).ne(1), -100
        )

        # not used during fine-tuning
        del batch["decoder_attention_mask"]

        # round down target lengths to multiple of reduction factor
        if model.config.reduction_factor > 1:
            target_lengths = torch.tensor(
                [len(feature["input_values"]) for feature in label_features]
            )
            target_lengths = target_lengths.new(
                [
                    length - length % model.config.reduction_factor
                    for length in target_lengths
                ]
            )
            max_length = max(target_lengths)
            batch["labels"] = batch["labels"][:, :max_length]

        # also add in the speaker embeddings
        batch["speaker_embeddings"] = torch.tensor(speaker_features)

        return batch
```

```python
import torch
from transformers import SpeechT5Processor, SpeechT5ForTextToSpeech, SpeechT5HifiGan
import soundfile as sf
from IPython.display import Audio
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Chemin de votre modèle
model_path = "/content/drive/MyDrive/Colab Notebooks/speecht5-darija-tts"

# Charger le modèle et processeur
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model = SpeechT5ForTextToSpeech.from_pretrained(model_path)
vocoder = SpeechT5HifiGan.from_pretrained("microsoft/speecht5_hifigan")

# Speaker embedding par défaut
speaker_embeddings = torch.randn(1, 512)

def test_tts(text, filename="test_output.wav", save_path="/content/drive/MyDrive/Colab Notebooks/"):
    """Test simple du modèle TTS avec sauvegarde dans un chemin spécifique"""

    # Créer le dossier s'il n'existe pas
    os.makedirs(save_path, exist_ok=True)

    # Chemin complet du fichier
    full_path = os.path.join(save_path, filename)

    # Convertir le texte
    inputs = processor(text=text, return_tensors="pt")

    # Générer l'audio
    with torch.no_grad():
        speech = model.generate_speech(inputs["input_ids"], speaker_embeddings, vocoder=vocoder)

    # Sauvegarder dans le chemin spécifié
    sf.write(full_path, speech.numpy(), samplerate=16000)
    print(f"✅ Audio sauvegardé dans: {full_path}")

    # Jouer l'audio
    return Audio(full_path, rate=16000)

# =======================================
# SYSTÈME D'ÉVALUATION
# =======================================

class DarijaTTSEvaluator:
    def __init__(self, audio_directory="/content/drive/MyDrive/Colab Notebooks/"):
        self.audio_dir = audio_directory

    def evaluate_audio_quality(self, filename):
        """Évaluation technique de la qualité audio"""
        filepath = os.path.join(self.audio_dir, filename)

        if not os.path.exists(filepath):
            print(f"❌ Fichier non trouvé: {filepath}")
            return None

        # Charger l'audio
        audio, sr = librosa.load(filepath, sr=16000)

        # 1. Métriques de base
        duration = len(audio) / sr
        rms_energy = np.sqrt(np.mean(audio**2))

        # 2. Détection de silence
        silence_threshold = 0.01
        silence_ratio = np.sum(np.abs(audio) < silence_threshold) / len(audio)

        # 3. Taux de passage par zéro (bruit)
        zcr = librosa.feature.zero_crossing_rate(audio)[0]
        avg_zcr = np.mean(zcr)

        # 4. Score de qualité (0-100)
        energy_score = min(rms_energy * 200, 40)
        silence_score = max(0, 30 - silence_ratio * 100)
        noise_score = max(0, 30 - avg_zcr * 1000)
        quality_score = min(energy_score + silence_score + noise_score, 100)

        return {
            'duration': duration,
            'rms_energy': rms_energy,
            'silence_ratio': silence_ratio,
            'zero_crossing_rate': avg_zcr,
            'quality_score': quality_score
        }

    def evaluate_pronunciation(self, filename, text):
        """Évaluation de la prononciation"""
        filepath = os.path.join(self.audio_dir, filename)

        if not os.path.exists(filepath):
            return None

        audio, sr = librosa.load(filepath, sr=16000)

        # Calculer vitesse de parole
        words = text.split()
        word_count = len(words)
        duration = len(audio) / sr
        words_per_second = word_count / duration if duration > 0 else 0

        # Score de vitesse (vitesse normale: 2-4 mots/sec)
        if 1.5 <= words_per_second <= 4.5:
            speed_score = 100
        elif words_per_second < 1.5:
            speed_score = 60  # Trop lent
        else:
            speed_score = 40  # Trop rapide

        # Évaluation de la vitesse
        if words_per_second < 1.5:
            speed_assessment = "Trop lent"
        elif words_per_second > 4.5:
            speed_assessment = "Trop rapide"
        elif 2.0 <= words_per_second <= 3.5:
            speed_assessment = "Parfait"
        else:
            speed_assessment = "Acceptable"

        return {
            'word_count': word_count,
            'duration': duration,
            'words_per_second': words_per_second,
            'speed_score': speed_score,
            'speed_assessment': speed_assessment
        }

    def quick_evaluation(self, filename, text):
        """Évaluation rapide d'un fichier"""
        print(f"\n🎤 ÉVALUATION: {filename}")
        print("-" * 40)

        # Évaluation technique
        tech = self.evaluate_audio_quality(filename)
        if tech:
            print(f"🔧 Qualité audio: {tech['quality_score']:.1f}/100")
            print(f"⏱️  Durée: {tech['duration']:.2f}s")
            print(f"🔊 Énergie: {tech['rms_energy']:.4f}")
            print(f"🤫 Silence: {tech['silence_ratio']*100:.1f}%")

        # Évaluation prononciation
        pron = self.evaluate_pronunciation(filename, text)
        if pron:
            print(f"🗣️  Vitesse: {pron['words_per_second']:.2f} mots/sec")
            print(f"📊 Évaluation vitesse: {pron['speed_assessment']}")
            print(f"⭐ Score vitesse: {pron['speed_score']}/100")

        # Score global
        if tech and pron:
            global_score = (tech['quality_score'] + pron['speed_score']) / 2
            grade = self.get_grade(global_score)
            print(f"\n🎯 SCORE GLOBAL: {global_score:.1f}/100")
            print(f"🏆 ÉVALUATION: {grade}")
            return global_score

        return 0

    def get_grade(self, score):
        """Convertir le score en note"""
        if score >= 85:
            return "Excellent 🌟"
        elif score >= 70:
            return "Très bien 👍"
        elif score >= 55:
            return "Bien 👌"
        elif score >= 40:
            return "Acceptable 🤔"
        else:
            return "À améliorer 📈"

# =======================================
# TESTS AVEC ÉVALUATION
# =======================================

# Créer l'évaluateur
evaluator = DarijaTTSEvaluator("/content/drive/MyDrive/Colab Notebooks/")

# Test simple
print("🎤 GÉNÉRATION ET ÉVALUATION TTS DARIJA")
print("=" * 50)

# Votre texte de test
test_text = "salam alikoum, Bghit nmchi lssouk"
filename = "test_darija_evaluation.wav"

# Générer l'audio
print("🔄 Génération de l'audio...")
audio_result = test_tts(test_text, filename)

# Jouer l'audio
display(audio_result)

# Évaluer le résultat
score = evaluator.quick_evaluation(filename, test_text)

# Conseils d'amélioration
print(f"\n💡 CONSEILS D'AMÉLIORATION:")
if score < 70:
    print("• Essayez différents speaker embeddings")
    print("• Vérifiez la qualité de vos données d'entraînement")
    print("• Augmentez le nombre d'epochs si possible")
else:
    print("• Votre modèle fonctionne bien!")
    print("• Testez avec des phrases plus longues")
    print("• Essayez différents types de voix")

print(f"\n🎉 Évaluation terminée!")
```
